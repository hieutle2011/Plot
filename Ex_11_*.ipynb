# Ex11_2

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pai = np.pi
x = np.arange(-2 * pai, 2 * pai, 0.3)
```

```python
matplotlib.style.use('ggplot')
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))
plt.legend(['sin', 'cos', 'tan'])
```
